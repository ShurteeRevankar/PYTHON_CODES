```python
# 1. summary with info

print("Patient Information")
print()
print(pinfo.info())
```

```python
# 2. Selecting relevant columns for billing

billing_cols = ['PatientID', 'Department', 'Doctor', 'BillAmount']
pinfo1 = pinfo[billing_cols]
pinfo1
```

```python
# 3. Droping administrative columns

pinfo = pinfo.drop(columns = ['ReceptionistID', 'CheckInTime'])
pinfo
```

```python
# 4. Total bill amount per department

dept_bill = pinfo.groupby('Department', as_index=False)['BillAmount'].sum()
print(dept_bill)
```

```python
# 5. Removing duplicate patient records

pinfo = pinfo.drop_duplicates(subset='PatientID')
pinfo
```

```python
# 6. Filling missing BillAmount values with mean amount 

pinfo['BillAmount'] = pd.to_numeric(pinfo['BillAmount'], errors='coerce')
mean_amt = pinfo.loc[pinfo['BillAmount'] != 0, 'BillAmount'].mean()
pinfo['BillAmount'] = pinfo['BillAmount'].fillna( mean_amt)
print(pinfo)
```

```python
# 7. Merge billing dataset with patient dataset

merged = pd.merge(pinfo, bill, on = 'PatientID', how = 'inner')
merged.head()
```

```python
# 8. Concatinating new data frame with new patients

new_patients = pd.DataFrame({
    'PatientID': [106, 107, 108, 109, 110],
    'Name' : ['Gracy', 'Shrutee', 'Brian', 'Alex', 'Stacy'],
    'Department': ['Cardiology', 'dermatology', 'Neurology', 'orthopedics', 'dermatology'],
    'Doctor': ['Dr. Smith', 'Dr. Rose', 'Dr. John', 'Dr. Lee', 'Dr. Rose'],
    'BillAmount': [6000.00, 4500.00, 5500.00, 5300.00, 7300.00]
})

new_patients
```

```python
concat_1 = pd.concat([pinfo, new_patients], axis=0, ignore_index=True)
concat_1
```
