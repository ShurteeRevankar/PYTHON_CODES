### Exercise 1: Prime Numbers

```python
import numpy as np 
```

```python
num = int(input("Enter a number: "))

for i in range(2, num):
    if num % i == 0:
        print(num, "is not a prime number")
        break
else:
    if num > 1:
        print("is a prime number")
    else:
        print("is not a prime number")
```

### Exercise 2: Product of Random Numbers

```python
import random

a = random.randint(1, 10)
b = random.randint(1, 10)

print("What is", a, "x", b, "?")
answer = int(input("Enter your answer: "))

if answer == a * b:
    print("Correct!")
else:
    print("Wrong! The correct answer is", a * b)
```

### Exercise 3: Squares of Even/Odd Numbers

```python
# squares of even numbers between 100 and 200

for num in range(100, 201):
    if num % 2 == 0:  # Checks number is even or not
        print("No:", num, "Sq:", num ** 2)    # If number is even then it prints its square
```

### Exercise 4:
Problem Statement: Hospital Billing System

```python
# Hospital Billing System (Beginner Version)

# Hospital Charges
room_charge_per_day = 2000
dr_fee = 1500
lab_charge_per_test = 300

# Patient details
days = 4
tests = 3
medicine_charge = 2400

# Calculate charges
room_total = room_charge_per_day * days
lab_total = lab_charge_per_test * tests

total = room_total + dr_fee + lab_total + medicine_charge

# Check for discount
if total > 10000:
    discount = total * 0.10
    total = total - discount
else:
    discount = 0

# Print final bill
print("Room Charges:", room_total)
print("Doctor Consultation Fee:", dr_fee)
print("Lab Test Charges:", lab_total)
print("Medicine Charges:", medicine_charge)
print("Discount Given:", discount)
print("Final Bill Total:", total)
```
